```python
#in this case we can replace 0 with 38.9 but we cant do that using Fillna function we have to create function
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass==1:
            return 38.23
        elif Pclass==2:
            return 29.87
        else:
            return 25.9
    else:
        return Age 
```

```python
titanic_data["Age"] = titanic_data[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
titanic_data.drop(['Name','Ticket','PassengerId'],axis=1,inplace=True)
```

```python
#MODEL BUILDING
```

```python
X = titanic_data.drop('Survived',axis=1)
y = titanic_data['Survived']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
model = LogisticRegression()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=20)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred =model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```
